Question 9. K-means clustering: Using the dataset in A3_P1.csv, implement K-means clustering and determine the number of clusters using the Elbow method.

```python
from google.colab import files

uploaded_files = files.upload()
```

```python
data = pd.read_csv("A3_P1.csv")
```

a. Plot the inertia (Within-Cluster Sum of Squares - WCSS) for number of cluster ranging from 1 to 15.

```python
inertia = []

# Cluster from 1 to 15
for num in range(1, 16):
    model = KMeans(n_clusters=num, random_state=42, n_init=10)
    model.fit(data)
    inertia.append(model.inertia_)

# Plotting the inertia values
plt.figure(figsize=(8, 5))
plt.plot(range(1, 16), inertia, marker='o', color='green', linestyle='-')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (WCSS)")
plt.title("Elbow Method for Optimal Clusters")
plt.show()
```

b. Find the optimal number of clusters using the elbow method.

```python
optimal_k = 4
```

c. Perform clustering using the optimal number of clusters, plot the clustering results with each cluster data in a different colour, and highlight the cluster centres.

```python
# Clustering using KMeans
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(data)

# Plotting the clusters
plt.figure(figsize=(8, 5))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=data['Cluster'], cmap='rainbow', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='black', marker='*', s=200, label="Centroids", edgecolor='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title(f"K-Means Clustering with k={optimal_k}")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

10. Hierarchical Clustering: Using the dataset in A3_P2.csv, implement bottom-up hierarchical clustering from scratch using Euclidean distance as the distance metric. Compute the distance between two clusters using the following methods:

a. 𝐷min(𝐴, 𝐵) = min 𝑢∈A, 𝑣∈B {𝛿(𝑢, 𝑣)} single-linkage clustering

b. 𝐷avg(𝐴, 𝐵) = ⟨𝛿(𝑢, 𝑣)⟩ 𝑢∈A, 𝑣∈B average-linkage clustering

c. 𝐷max(𝐴, 𝐵) = max 𝑢∈A, 𝑣∈B {𝛿(𝑢, 𝑣)} complete-linkage clustering.

Plot dendrograms for each clustering method to visualize the hierarchical clustering process.

```python
from google.colab import files

uploaded_files = files.upload()
```

```python
data = pd.read_csv('A3_P2.csv')
X = data.values  # Converting to numpy array
```

```python
# Single Linkage Dendrogram

Z = sch.linkage(X, method = 'single')
plt.figure(figsize=(10,4))
den = sch.dendrogram(Z)
plt.title('Single Linkage Dendogram')
plt.xlabel('Data Points Number')
plt.ylabel('Euclidean distance')
```

```python
# Average Linkage Dendrogram

Z = sch.linkage(X, method = 'average')
plt.figure(figsize=(10,4))
den = sch.dendrogram(Z)
plt.title('Average Linkage Dendrogram')
plt.xlabel('Data Points Number')
plt.ylabel('Euclidean distance')
```

```python
# Complete Linkage Dendrogram

Z = sch.linkage(X, method = 'complete')
plt.figure(figsize=(10,4))
den = sch.dendrogram(Z)
plt.title('Complete Linkage Dendrogram')
plt.xlabel('Data Points Number')
plt.ylabel('Euclidean distance')
```
